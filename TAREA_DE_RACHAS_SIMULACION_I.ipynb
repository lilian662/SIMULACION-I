## Clasificación de los números

Los números se clasifican como:

- "+" si el número es mayor o igual a 0.5  
- "-" si el número es menor a 0.5

```python
clasificacion = []

for x in datos:
    if x >= 0.5:
        clasificacion.append("+")
    else:
        clasificacion.append("-")

print("Clasificación:")
print(clasificacion)
```

## Cálculo del número de rachas

```python
rachas = 1

for i in range(1, n):
    if clasificacion[i] != clasificacion[i-1]:
        rachas += 1

print("Número de rachas:", rachas)
```

## Conteo de símbolos

```python
n1 = clasificacion.count("+")
n2 = clasificacion.count("-")

print("Cantidad de + :", n1)
print("Cantidad de - :", n2)
```

## Cálculo de media y varianza

```python
media = ((2*n1*n2) / (n1+n2)) + 1

varianza = (2*n1*n2*(2*n1*n2 - n1 - n2)) / (((n1+n2)**2)*(n1+n2 - 1))

desv = sqrt(varianza)

print("Media esperada:", media)
print("Varianza:", varianza)
```

## Cálculo del estadístico Z

```python
Z = (rachas - media) / desv

print("Estadístico Z:", Z)
```

## Conclusión de la prueba

```python
if abs(Z) < 1.96:
    print("No se rechaza H0: la secuencia es independiente.")
else:
    print("Se rechaza H0: la secuencia no es independiente.")
```
