## 1. Fundamento Teórico

La prueba Kolmogorov–Smirnov (K–S) permite comparar una función de distribución empírica
con una función de distribución teórica.

Se desea contrastar:

H₀: Los datos siguen una distribución Uniforme U(0,1)  
H₁: Los datos no siguen una distribución Uniforme U(0,1)

El estadístico de prueba es:

$$
D_n = \sup_x |F_n(x) - F(x)|
$$

donde:

- $F_n(x)$ es la función de distribución empírica.
- $F(x)$ es la función de distribución teórica.

Para la distribución Uniforme U(0,1):

$$
F(x) = x \quad \text{para } 0 \le x \le 1
$$

El valor crítico se calcula como para datos mayores a 100:

$$
D_{\alpha} = \frac{1.36}{\sqrt{n}} \quad \text{para } \alpha = 0.05
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [
0.78961, 0.05230, 0.10699, 0.55877, 0.14151,
0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
0.80548, 0.82654, 0.29453, 0.20852, 0.42989,
0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
0.89898, 0.57880, 0.67621, 0.05010, 0.00121,
0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
0.38618, 0.76910, 0.68334, 0.55170, 0.10850,
0.79882, 0.45679, 0.21631, 0.87616, 0.55743,
0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
0.29931, 0.30861, 0.83358, 0.51781, 0.03272,
0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
0.24000, 0.65559, 0.38507, 0.90829, 0.94187,
0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
0.54325, 0.62400, 0.09133, 0.41678, 0.33954,
0.58244, 0.85853, 0.86752, 0.38729, 0.15506,
0.23949, 0.53559, 0.33381, 0.49883, 0.75103,
0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
0.19147, 0.40644, 0.08128, 0.73435, 0.22724,
0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

datos = np.array(data)
n = len(datos)

print("Tamaño de muestra:", n)
```

```python
datos_ordenados = np.sort(datos)

datos_ordenados
```

```python
Fn = np.arange(1, n+1) / n
```

```python
# Cálculo de D+
D_plus = np.max(Fn - datos_ordenados)

# Cálculo de D-
D_minus = np.max(datos_ordenados - (np.arange(0, n) / n))

# Estadístico final
Dn = max(D_plus, D_minus)

print("D+ =", D_plus)
print("D- =", D_minus)
print("Dn =", Dn)
```

```python
D_critico = 1.36 / np.sqrt(n)

print("D crítico =", D_critico)
```

```python
if Dn > D_critico:
    print("Se rechaza H0: Los datos NO siguen U(0,1)")
else:
    print("No se rechaza H0: Los datos siguen U(0,1)")
```

```python
plt.figure(figsize=(8,6))

plt.step(datos_ordenados, Fn, where='post', label='F_n(x) Empírica')
x = np.linspace(0,1,100)
plt.plot(x, x, label='F(x)=x Uniforme U(0,1)')

plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Prueba Kolmogorov-Smirnov")
plt.legend()
plt.grid()
plt.show()
```
